## Step 1 - get some data in.

```python
import pandas as pd

# the data is from a web site purporting to list all draws since 1994 in the order the numbers came out

df = pd.read_csv('/home/chris/Documents/lottery2.csv')

df.columns = ('Draw', 'Day', 'Date', 'Month', 'Year', 'N1', 'N2', 'N3', 'N4', 'N5', 'N6', 'B', 'Jackpot', 'Wins', 'Machine', 'Set')

df = df[['N1','N2','N3','N4','N5','N6','B']] 

df

```

## Just calculating the sum of each row

```python
df['Sum'] = df.sum(axis=1)

df[['Sum']]
```

## multiply each number by each othe number in each row

```python
df['Mul'] = df[['N1']].multiply(df['N2'].multiply(df['N3']).multiply(df['N4']).multiply(df['N5']).multiply(df['N6']).multiply(df['B']), axis='index')
df[['Mul']]
```

## Divide Mul by Sum, subtract Divres from Mul (Diff)

```python
import pandas as pd
import numpy as np

z = []

df['Divres'] = df['Mul']/df['Sum']

df['Diff'] = df['Mul']-df['Divres']

df['Divres'] = np.round(df[['Divres']], 0)

df['Diff'] = np.round(df[['Diff']], 0)

df['Divres'] = df['Divres'].astype(str)

df[['Sum','Mul','Divres', 'Diff']]

for i in df.Diff:
    
    z.append[i]
    #str = z[i]

    print z
```
